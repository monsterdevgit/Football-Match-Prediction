```python
matches["date"] = pd.to_datetime(matches["date"])
```

```python
matches["target"] = (matches["result"] == "W").astype("int")     #This assigns the resulting integer values to a new column named "target". Each element in the "target" column will be 1 if the corresponding match result is "W" (win) and 0 otherwise.
```

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes   #It assigns a numerical code to each unique value in the "venue" column.
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes   #It assigns a numerical code to each unique value in the "opponent" column.
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")    #It extracts the hour part from the "time" column 
```

```python
matches["day_code"] = matches["date"].dt.dayofweek  #It extracts the day of the week from the "date" column
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'min_samples_split': [2, 5, 10],
    'max_depth': [None, 5, 10]
}

# Create the random forest classifier
rf = RandomForestClassifier(random_state=1)
```

```python
train = matches[matches["date"] < '2023-01-01']
test = matches[matches["date"] > '2023-01-01']
```

```python
# Calculate the percentage of train and test data
train_percent = (len(train) / len(matches)) * 100
test_percent = (len(test) / len(matches)) * 100

# Print the percentages
print("Train Data Percentage: {:.2f}%".format(train_percent))
print("Test Data Percentage: {:.2f}%".format(test_percent))     #studies show that the best results are obtained if we use 20-30% of the data for testing, and the remaining 70-80% of the data for training.
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
# Perform grid search
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5)
grid_search.fit(train[predictors], train["target"])

# Print the best parameters and best score
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```

```python
# Access the best model from grid search
best_model = grid_search.best_estimator_

# Make predictions using the best model
preds = best_model.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
error = accuracy_score(test["target"], preds)
error
```

```python
combined = pd.DataFrame(dict(actual=test["target"], predicted=preds))
```

```python
pd.crosstab(index=combined["actual"], columns=combined["predicted"])
```

```python
from sklearn.metrics import precision_score

precision_score(test["target"], preds)    #model correctly identifies around 51.43% of the positive predictions.
```

**Another Feature Engineering for Model**

```python
grouped_matches = matches.groupby("team")
group = grouped_matches.get_group("Manchester City").sort_values("date")  #groups the "matches" DF by team, selects a specific team's group
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group                   ##calculates rolling averages for specified columns, adds the rolling averages as new columns, removes rows with missing rolling average values, and returns the modified DataFrame.
```

```python
cols = ["gf", "ga","xg"]    #, "fk", "pk", "pkatt"
new_cols = [f"{c}_rolling" for c in cols]

rolling_averages(group, cols, new_cols) 
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
matches_rolling                       #takes the rolling_averages function and applies to each team
```

```python
matches_rolling = matches_rolling.droplevel('team')
matches_rolling          #removes team as index
```

```python
predictors_new= ["venue_code", "opp_code", "hour", "day_code", "sh", "sot"]
predictors_new
```

```python
def make_predictions(data, predictors_new):
    train = data[data["date"] < '2023-01-01']
    test = data[data["date"] > '2023-01-01']
    
    grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5)
    grid_search.fit(train[predictors_new], train["target"])
    best_model = grid_search.best_estimator_
   
    preds = best_model.predict(test[predictors_new])
    
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    error = precision_score(test["target"], preds)
    return combined, error
```

```python
combined, error = make_predictions(matches_rolling, predictors_new + new_cols)
error
```

```python
combined = combined.merge(matches_rolling[["date", "team", "opponent", "result"]], left_index=True, right_index=True)

#combined and matches_rolling are merged based on their index values. The resulting DataFrame contains columns from both DataFrames,
```

```python
class MissingDict(dict):
    __missing__ = lambda self, key: key

map_values = {"Brighton and Hove Albion": "Brighton", "Manchester United": "Manchester Utd", "Newcastle United": "Newcastle Utd", "Tottenham Hotspur": "Tottenham", "West Ham United": "West Ham", "Wolverhampton Wanderers": "Wolves"} 
mapping = MissingDict(**map_values)
```

```python
combined["new_team"] = combined["team"].map(mapping)
merged = combined.merge(combined, left_on=["date", "new_team"], right_on=["date", "opponent"])
```
